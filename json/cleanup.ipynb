```python
import pandas as pd, urllib, requests, urllib2, json, numpy as np
import zipfile
import os
```

```python
from os import listdir
from os.path import isfile, join
working_folder='./'
b = [i for i in set([f for f in listdir(working_folder) if not isfile(join(working_folder, f))]) if i[0]!='.']
```

```python
c={i: listdir(working_folder+i) for i in b}
```

```python
for z in c:
    countries=set([i[:-6] for i in c[z] if i[-6:-5]=='m'])
    for x in countries:
        files=[i for i in c[z] if (x in i)]
        ZipFile = zipfile.ZipFile(x+".zip", "w" )
        for a in files:
            ZipFile.write(working_folder+'/'+a[0]+'/'+a, arcname=a, compress_type=zipfile.ZIP_DEFLATED)
        ZipFile.close()
        print x
```
